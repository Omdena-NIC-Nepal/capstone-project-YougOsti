```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline  

# Load cleaned data
df = pd.read_csv(r"C:\Users\youg_\capstone-project-YougOsti\Data\Raw\Environment_data\number-of-threatend-spicies-1998_2018table06_NSO.csv", low_memory=False)
```

```python
df.sample(4)
```

```python
# STEP 1: Remove unnamed columns or extra index columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# STEP 2: Rename first column to something meaningful if it's not already
df.rename(columns={df.columns[0]: "Species"}, inplace=True)

# STEP 3: Convert year columns to numeric
df.columns = [str(col).strip() for col in df.columns]  # clean column names
for col in df.columns[1:]:
    try:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    except:
        pass

# STEP 4: Handle missing values (forward then backward fill across rows)
df = df.set_index("Species")
df = df.ffill(axis=1).bfill(axis=1).reset_index()

# STEP 5: Check the result
print(df.info())
print(df.head())
```

```python
# Define the output path
output_path = r"C:\Users\youg_\capstone-project-YougOsti\Data\processed\threatened_species_cleaned.csv"

# Save the cleaned DataFrame
df.to_csv(output_path, index=False)
```
